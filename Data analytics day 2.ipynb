```python
#merging two data frames
df3=pd.DataFrame({'int_rate':[2,1,2,3],'int_gdp':[50,45,45,67]},index=[1997,1998,1999,2000])
df3
```

```python
df4=pd.DataFrame({'low_tier_hpi':[80,90,70,60],'unemployment':[1,3,5,6]},index=[2001,2002,2003,2004])
df4
```

```python
df3.join(df4)# arranged based upon index values
```

```python
#merging the dataframes having similar column values
m=pd.DataFrame({'a':[4,5,6,7],'b':[30,50,60,70],'c':['a','b','c','d']},index=['first','second','third','four'])
m
```

```python
n=pd.DataFrame({'a':[4,8,6,7],'b':[30,56,60,70],'c':['a','b','c','d']},index=['first','second','third','four'])
n
```

```python
m.merge(n)
```

```python
new=m.merge(n)
new.columns=['f','s','th']
new.index=[9,7,6]
new
```

```python
m.merge(n,on="c")
```

```python
n.merge(m,on='a')
```
