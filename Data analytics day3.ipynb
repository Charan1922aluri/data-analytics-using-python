```python
from matplotlib import style 
style.use("ggplot")
days=[1,2,3,4,5]
temp=[36,35,37.7,38,40.2]
plt.plot(days,temp,color='b',linestyle='dashed',marker=".",linewidth=2,markersize=10)
plt.xlabel('Days',fontsize="10")
plt.ylabel("Temperature",fontsize="10")
plt.title("Day vs Temperature graph")
plt.show()
```

```python
vij_temp=[40.5,43.5,38.5,39.2,39.5]
plt.plot(days,vij_temp,'b>--',label='vij')
plt.plot(days,temp,'m*--',label='gnt')
plt.xlabel('date')
plt.ylabel('temperature')
plt.title("vij vs gnt temperatures")
plt.legend(loc=4)
plt.grid(color='red',linestyle='-',linewidth=2)
plt.show()
```

### Histograms
- represents a bin
- bins represent the data  points in th range
- plt.hist(ele,bins=count)

```python
std1=np.random.randint(18,45,100)
std2=np.random.randint(15,40,100)
plt.hist(std1,bins=7)
plt.xlabel('age category')
plt.ylabel('no of studens')
plt.show()
```

```python
plt.hist(std2,bins=10,color='c')
plt.show()
```

```python
plt.hist(std2,bins=10,color='c',orientation='horizontal')
plt.show()
```

```python
vals=[num for num in range(10,50,5)]
plt.hist(std1,vals,histtype='bar',orientation='vertical',color='orange')
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(8,5))
plt.hist([std1,std2],vals,rwidth=0.7,histtype='bar',orientation='vertical',color=['y','c'])
plt.title("multiple histograms")
plt.show()
```

### pie chart

```python
f=['good','average','excellent','modrate','poor']
plt.pie(x,labels=f,shadow=True,colors=['red','orange','grey','pink','cyan'],autopct="%.f%%",startangle=45,explode=[1,1,0,0,0])
plt.title('sort by Responses')
plt.show()
```
